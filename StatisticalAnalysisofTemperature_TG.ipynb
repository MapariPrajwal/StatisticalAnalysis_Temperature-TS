```python
Df.rename(columns={"Min Temp (°C)":"Min_Temp"},inplace=True)
Df.rename(columns={"Max Temp (°C)":"Max_Temp"},inplace=True)
```

```python
def avg_temp(row):
  return (row.Min_Temp+row.Max_Temp)/2
```

```python
Df['Avg_Temp'] = Df.apply(avg_temp,axis=1)
```

```python
drop_cols=[5]
Df.drop(Df.columns[drop_cols],axis=1,inplace=True)
```

```python
Df.set_index('Date',inplace=True)
```
